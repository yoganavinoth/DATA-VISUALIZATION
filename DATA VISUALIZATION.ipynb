# distplot

A dist plot is a type of data visualization plot used to represent the distribution of a numerical variable. It shows the shape of the distribution, central tendency, and variability of the data. The plot essentially creates a histogram with a density curve drawn on top of it, which provides a smooth estimate of the distribution. The dist plot is a useful tool for exploring and understanding the distribution of a dataset, identifying any outliers or skewness, and determining the most appropriate statistical analysis to perform on the data. It can be easily created using the distplot() function in Seaborn library in Python.

```python
import seaborn as sb
sb.distplot(dataset["ssc_p"],hist=True,kde=True,kde_kws={"color":"purple"})
```

# jointplot

A joint plot is a type of data visualization that combines two different plots into one figure. It is typically used to display the relationship between two variables, such as height and weight, or age and income. 


A joint plot can be created using the Seaborn library in Python. The most common type of joint plot is the scatter plot, which displays the data as a set of points. Other types of joint plots include hex plots and contour plots, which use different techniques to display the data.


A joint plot can be useful for identifying trends and patterns in the data, as well as for detecting outliers and other anomalies. It can also be used to compare different groups or data sets, and to explore the relationships between multiple variables.

```python
sb.jointplot(x="ssc_p",y="salary",data=dataset)
plt.show
```

```python
sb.jointplot(x="ssc_p",y="salary",data=dataset,kind="hex")
plt.show
```

```python
sb.jointplot(x="ssc_p",y="salary",data=dataset,kind="kde")
plt.show
```

# stripplot

stripplot-categorical plot .A strip plot is a type of plot used to display the distribution of a dataset by using points to represent individual observations. It is similar to a scatter plot, but with the points adjusted along one axis so they do not overlap. The stripplot() function in Seaborn allows for the creation of a categorical scatterplot, where one variable is categorical. It can be used to show the distribution of values but is not ideal for large numbers of observations. It is also a good complement to box or violin plots

# violinplot

A violin plot is a type of plot used to display the distribution of a dataset by using density curves, similar to kernel density plots. The width of each curve reflects the approximate frequency of data points in that region. It allows for the comparison of distributions between multiple groups and can be combined with other chart types, such as box plots, to provide additional information. Violin plots are suitable for situations where the dataset is too big to show individual observations and are better than box plots since they provide more information.

```python
#Distribution Observation
sb.violinplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

```python
sb.violinplot(x="specialisation",y="salary",data=dataset,hue="gender")
plt.show()
```

# boxplot

A box plot, also known as a box and whisker plot, is a graphical representation of the distribution of a dataset based on a five-number summary. [1][2] The five-number summary includes the minimum, first quartile, median, third quartile, and maximum values of the dataset. The box in the middle of the plot represents the interquartile range (IQR), which is the distance between the first and third quartiles. The whiskers extend from the box to the smallest and largest values within 1.5 times the IQR from the box. Any data point beyond the whiskers is considered an outlier. Box plots are useful for comparing the distribution of data between different groups or data sets.

```python
#boxplot-means percentile
sb.boxplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender")
plt.show()
```

# Barplot

A bar plot is a type of plot that displays categorical data with rectangular bars. The height or length of each bar represents the frequency or count of data in that category. Bar plots are useful for comparing the frequency of different categories, identifying trends in data, and showing differences between groups. They are commonly used in business, economics, and social sciences to present survey results, market shares, and other categorical data. Bar plots can be created using various software tools, such as Excel, Python's Matplotlib library.

```python
#Barplot
sb.barplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender");
plt.show()
```

# countplot

A count plot is a type of plot used to display the counts of observations in each categorical bin using bars [1][2]. It can be thought of as a categorical histogram, where the variable on the x-axis is categorical instead of quantitative. The countplot() function in Seaborn is commonly used to create count plots and it treats one of the variables as categorical even when the data has a numeric or date type [1]. Count plots are useful to show the frequency or count of data in different categories, and they are commonly used in business, economics, and social sciences. To create a count plot, one can specify the variables for data, x, y, hue, order, hue_order, orient, color, palette, saturation, width, dodge, and ax using the countplot() function [1].



```python
#countplot
sb.countplot(x="ssc_b",data=dataset,hue="gender")
plt.show()
```

# pointplot

A point plot is a type of categorical plot that shows the mean (or another estimator) of a dependent variable for each categorical level using points. It is typically used to compare two or more groups across one or more independent variables. Point plots are useful for illustrating differences in means or other measures of central tendency between groups. They can also be used to display confidence intervals or standard errors around the means. Point plots can be created using various data visualization libraries in Python, such as Seaborn or Matplotlib.

```python
#pointplot
sb.pointplot(x="hsc_p",y="hsc_b",data=dataset,hue="gender")
plt.show()
```

# factorplot

factorplot is a function in the Seaborn library that allows you to create different types of categorical plots like barplot, countplot, pointplot, etc. The factorplot function helps to visualize the distribution of a variable across different levels of one or more categorical variables. It can also be used to display the relationship between two variables, where one is categorical and the other is numeric. factorplot is useful in analyzing complex datasets with multiple variables and can be customized using various parameters like kind, size, aspect, col, row, hue, etc.

```python
#factorplot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("exercise")
sb.factorplot(x="time",y="pulse",hue="kind",kind="violin",col="diet",data=df);
plt.show()
```
